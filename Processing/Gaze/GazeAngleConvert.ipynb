```python
import pandas as pd
import numpy as np
import os             
import math
date="03-12-2022"

directory="../3people/"+date+"/Gaze/"
all_files = os.listdir(directory)
txt_files = filter(lambda x: x[-4:] == '.txt', all_files)
output_directory="../3people/"+date+"/Gaze/Convert/"
# list(txt_files)
all_files=list(txt_files)
def isNaN(num):
    return num != num
```

```python
z=((960/np.tan(np.deg2rad(46)))+(540/np.tan(np.deg2rad(33.5))))/2

for file_name in all_files:
    colnames=['X', 'Y'] 
    df = pd.read_csv(directory+file_name, names=colnames, header=None)
    
#     if math.isnan(df['X'][0]):
#         df['beta']="Nan"
#         df['alpha']= "Nan"
#     else:
        
    df['X']=df['X']-960
    df['Y']=540 - df['Y']
    df['sqrt']=(df['X'] ** 2+df['Y'] ** 2+z ** 2) ** 0.5
    df['beta']=np.rad2deg(np.arcsin(df['Y']/df['sqrt']))
    df['alpha']= np.rad2deg(np.arcsin((df['X']/df['sqrt'])/np.cos(np.deg2rad(df['beta']))))
    new = df[['alpha','beta']]
    new.to_csv(output_directory+file_name, sep=' ', index=False,na_rep='NaN', header=False)


```
